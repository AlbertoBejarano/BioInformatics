```python
import gzip
import pandas as pd
import warnings
warnings.filterwarnings("ignore") 
```

```python
expression_link = 'http://gdac.broadinstitute.org/runs/stddata__2016_01_28/data/ACC/20160128/gdac.broadinstitute.org_ACC.Merge_rnaseqv2__illuminahiseq_rnaseqv2__unc_edu__Level_3__RSEM_genes_normalized__data.Level_3.2016012800.0.0.tar.gz'
```

```python
df = pd.read_csv(expression_link, compression='gzip', header=0, sep='\t', quotechar='"', error_bad_lines=False)
```

```python
#%%capture
df0 = df.iloc[:, 0:5]
df0.columns.tolist()
```

```python
df.head()
```

```python
df.drop(0, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df['location'] = df['././@LongLink'].astype(str)
```

```python
df['GeneID'] = df['location'].str.split('|',expand=False)
df['GeneID'] = df['location'].str.split('|',expand=True)
```

```python
first_column = df.pop('GeneID')
df.insert(0, 'GeneID', first_column)
```

```python
df.drop(['././@LongLink', 'location'], axis=1, inplace=True)
```

```python
df = df.dropna()
df.reset_index(drop=True, inplace=True)
```

```python
df = df[~df.isin(['?']).any(axis=1)]
df.reset_index(drop=True, inplace=True)
```

```python
df.to_csv("./Data/Output/OutputPySingscrore.txt", sep="\t", index=False)
```
