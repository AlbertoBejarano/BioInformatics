```python
from scipy import stats
import pandas as pd
```

```python
pd.options.display.float_format = '{:.3f}'.format
```

```python
GeneSign_df = pd.read_csv("./data/GeneSignatures.tsv", sep='\t')
GeneExp_df  = pd.read_csv("./data/GeneExpression.tsv", sep='\t')
```

```python
GeneSign_df0 = GeneSign_df.melt(id_vars = None, var_name = 'GeneSignature', value_name = 'Gene')
```

```python
GeneExp_df = GeneExp_df.replace(0,0.001)
#GeneExp.drop('Sample', axis=1,inplace=True)
GeneExp_df.head(3)
```

```python
signatures = GeneSign_df0.GeneSignature.unique().tolist()
signatures.sort(reverse=True)
```

```python
df00 = []
#     -     -     -     -     -     -     -     -     -     -     -     -     -     -     -     -
for sign in signatures:
    df00 = GeneSign_df0[GeneSign_df0.GeneSignature == sign]
    GeneList = df00.Gene.tolist()
    CleanedGeneList = [x for x in GeneList if str(x) != 'nan']
    #GeneExp_df[b] = GeneExp_df[CleanedGeneList].mean(axis=1, numeric_only=True)
    GeneExp_df[sign] = stats.gmean(GeneExp_df[CleanedGeneList],axis=1)
    first_column = GeneExp_df.pop(sign); GeneExp_df.insert(1, sign, first_column)
    print(sign, "\t", CleanedGeneList)
```
